### Importing Required Dependencies <a name = "library"></a>

```python
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Input
from keras.models import Model
from keras.utils import plot_model
```

### ALexNet Model Function <a name = "model_function"></a>

```python
from os import name
def AlexNet():
  input_layer = Input(shape = (227,227,3), name = 'input_layer')
  # Block_1
  x = Conv2D(filters = 96, kernel_size = (11,11), strides = (4,4), activation = 'relu', padding = 'valid', name = 'block1_conv1')(input_layer)
  x = MaxPooling2D((3,3), strides = (2,2), name = 'block1_pool')(x)
  # Block_2
  x = Conv2D(filters = 256, kernel_size = (5,5), strides = (1,1), activation = 'relu', padding = 'same', name = 'block2_conv1')(x)
  x = MaxPooling2D((3,3), strides = (2,2), name = 'block2_pool')(x)
  # Block_3
  x = Conv2D(filters = 384, kernel_size = (3,3), strides = (1,1), activation = 'relu', padding = 'same', name = 'block3_conv1')(x)
  x = Conv2D(filters = 384, kernel_size = (3,3), strides = (1,1), activation = 'relu', padding = 'same', name = 'block3_conv2')(x)
  x = Conv2D(filters = 256, kernel_size = (3,3), strides = (1,1), activation = 'relu', padding = 'same', name = 'block3_conv3')(x)
  x = MaxPooling2D((3,3), strides = (2,2), name = 'block3_pool')(x)
  # Fully Connected Block
  x = Flatten()(x)
  x = Dense(4096, activation = 'relu', name = 'dense1')(x)
  x = Dense(4096, activation = 'relu', name = 'dense2')(x)
  output = Dense(1000, activation = 'softmax', name = 'output_layer')(x)

  model = Model(inputs = input_layer, outputs = output, name = 'AlexNet Model')

  return model
```

```python
model = AlexNet()
```

### Model Summary <a name = "summary"></a>

```python
model.summary()
```

### Plotting The Model Architecture <a name = "plot"></a>

```python
plot_model(model, show_shapes = True, show_layer_activations = True, show_layer_names = True, to_file = 'model.png')
```
